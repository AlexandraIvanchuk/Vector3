```python
SentenceVectors = {}
SentenceText = {}
FolderPath = 'Docs'
Files = ['50_00_2020.txt','50_01_2015.txt','50_04_2011.txt','50_07_2011.txt','50_08_2011.txt','50_09 _2011.txt','50_10_2011.txt','50_11_2011.txt','50_14_2011.txt','50_17_2015.txt','50_18_2015.txt']
```

```python
for FileName in Files:
    print(FileName)
    FilePath = f'{FolderPath}/{FileName}'
    with open(FilePath, 'r', encoding='utf-8') as File:
        Text = File.read()
        NlpText = nlp(Text)
        Sentences = list(NlpText.sents)
        for i, Sentence in enumerate(Sentences):
            NlpSentence = nlp(str(Sentence))
            FilteredSentence = " ".join([token.text for token in NlpSentence if not token.is_stop])
            NlpFilteredSentence = nlp(FilteredSentence)
            print(NlpFilteredSentence)
            FilteredVector = NlpFilteredSentence.vector
            SentenceVectors[f"{FileName}_sentence_{i+1}"] = FilteredVector
            print(FilteredVector)
            SentenceText[f"{FileName}_sentence_{i+1}"] = NlpFilteredSentence
```

```python
from scipy.spatial.distance import cosine

def MostSimilar(QueryVctor, SentenceVectors):
    MinDistanceValue = float('inf')
    MostSimilarSentence = None
    
    for Key, Vector in SentenceVectors.items():
        Distance = cosine(QueryVctor, Vector)
        
        if Distance < MinDistanceValue:  
            MinDistanceValue = Distance
            MostSimilarSentence = Key
    
    return MostSimilarSentence, MinDistanceValue

def MostSimilarity(NlpQuerySentence, SentenceText):
    MinDistanceValue = float('inf')
    MostSimilarSentence = None
    
    for Key, Sentence in SentenceText.items():
        Similar = NlpQuerySentence.similarity(Sentence)
        Distance = 1 - Similar
        
        if Distance < MinDistanceValue:  
            MinDistanceValue = Distance
            MostSimilarSentence = Key
    
    return MostSimilarSentence, MinDistanceValue

#из файла 50_10_2011
QuerySentence = "В соответствии с требованиями к данным, определённым на этапе анализа БД (см. 7.1), методикой (см. 7.2) и параметрами управления объёмами данных выполняется" 
NlpQuerySentence = nlp(QuerySentence)
QueryVctor = NlpQuerySentence.vector
MostSimilarSentence, MinDistanceValue = MostSimilar(QueryVctor, SentenceVectors)

print(f"Наиболее похожее предложение через cosine: {MostSimilarSentence} с косинусным расстоянием: {MinDistanceValue}")

MostSimilarSentence, MinDistanceValue = MostSimilarity(NlpQuerySentence, SentenceText)

print(f"Наиболее похожее предложение через similarity: {MostSimilarSentence} с косинусным расстоянием: {MinDistanceValue}")
```
